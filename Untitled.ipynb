```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import matplotlib.pyplot as plt
import pystac_client
import planetary_computer
import odc.stac
```

```python
LAT = -3.7327
LON = -45.3648
DATA1_INICIO = "2022-07-01"
DATA1_FIM    = "2022-09-30"
DATA2_INICIO = "2024-07-01"
DATA2_FIM    = "2024-09-30"
```

```python
delta = 0.15
bbox = [LON-delta, LAT-delta, LON+delta, LAT+delta]

catalog = pystac_client.Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1",
    modifier=planetary_computer.sign_inplace,
)

def buscar(d_ini, d_fim):
    search = catalog.search(
        collections=["sentinel-2-l2a"],
        bbox=bbox,
        datetime=f"{d_ini}/{d_fim}",
        query={"eo:cloud_cover": {"lt": 30}},
        sortby="eo:cloud_cover",
    )
    items = list(search.items())
    if not items:
        raise Exception(f"Nenhuma imagem encontrada para {d_ini} a {d_fim}")
    print(f"✅ Imagem encontrada: {items[0].id} | Nuvens: {items[0].properties['eo:cloud_cover']:.1f}%")
    return items[0]

item1 = buscar(DATA1_INICIO, DATA1_FIM)
item2 = buscar(DATA2_INICIO, DATA2_FIM)
```

```python
def calcular_ndvi(item):
    ds = odc.stac.load([item], bands=["B04","B08"], resolution=60, bbox=item.bbox)
    red = ds["B04"].isel(time=0).values.astype(float)
    nir = ds["B08"].isel(time=0).values.astype(float)
    with np.errstate(divide="ignore", invalid="ignore"):
        ndvi = np.where((nir+red)==0, np.nan, (nir-red)/(nir+red))
    return ndvi

ndvi1 = calcular_ndvi(item1)
ndvi2 = calcular_ndvi(item2)

h = min(ndvi1.shape[0], ndvi2.shape[0])
w = min(ndvi1.shape[1], ndvi2.shape[1])
ndvi1, ndvi2 = ndvi1[:h,:w], ndvi2[:h,:w]
ndvi_diff = ndvi1 - ndvi2
print("✅ NDVI calculado!")
```

```python
mask = ~np.isnan(ndvi_diff)
total = np.sum(mask)
pct_degradacao = 100 * np.sum((ndvi_diff > 0.25) & mask) / total
pct_melhora    = 100 * np.sum((ndvi_diff < -0.25) & mask) / total
delta_medio    = np.nanmean(ndvi_diff)

print("=" * 40)
print("📊 INSIGHTS GEO-SIGHT")
print("=" * 40)
print(f"🔴 Área degradada:   {pct_degradacao:.1f}%")
print(f"🟢 Área recuperada:  {pct_melhora:.1f}%")
print(f"📈 ΔNDVI médio:      {delta_medio:+.3f}")

if pct_degradacao > 20:
    print("⚠️  ALERTA CRÍTICO: degradação expressiva detectada!")
elif pct_degradacao > 10:
    print("🟡 ALERTA MODERADO: monitoramento recomendado.")
else:
    print("🟢 ESTÁVEL: sem degradação significativa.")

vmax = max(abs(np.nanpercentile(ndvi_diff, 2)), abs(np.nanpercentile(ndvi_diff, 98)), 0.1)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].imshow(ndvi1, cmap="RdYlGn", vmin=-1, vmax=1)
axes[0].set_title(f"NDVI — ANTES\n{item1.properties['datetime'][:10]}")
axes[0].axis("off")

axes[1].imshow(ndvi2, cmap="RdYlGn", vmin=-1, vmax=1)
axes[1].set_title(f"NDVI — DEPOIS\n{item2.properties['datetime'][:10]}")
axes[1].axis("off")

im = axes[2].imshow(ndvi_diff, cmap="RdYlGn", vmin=-vmax, vmax=vmax)
axes[2].set_title("ΔNDVI (Antes − Depois)\n🔴 Degradação | 🟢 Recuperação")
axes[2].axis("off")
plt.colorbar(im, ax=axes[2], fraction=0.03)

plt.suptitle("🛰️ Geo-Sight — Detecção de Mudanças Ambientais", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("geo_sight_resultado.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Mapa salvo como geo_sight_resultado.png")
```
